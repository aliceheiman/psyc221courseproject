```python
income_features = [("demo_comb_income_v2_" + str(i) + ".0").strip() for i in range(1, 11) if i != 9]
demo_features = income_features + ["demo_race_black", "demo_race_other"]

def one_hot_encode(input_df):
    df = input_df.copy();
    df["demo_comb_income_v2"] = df["demo_comb_income_v2"].replace({999: None, 777: None})
    df = pd.get_dummies(df, columns=['demo_comb_income_v2'], drop_first=False)

    df = df.drop("demo_comb_income_v2_9.0", axis=1) # most common
    income_features = [("demo_comb_income_v2_" + str(i) + ".0").strip() for i in range(1, 11) if i != 9]
    
    df["demo_race_black"] = df["demo_race_a_p___11"] == 1
    df["demo_race_other"] = (df["demo_race_a_p___10"] != 1) & (df["demo_race_a_p___11"] != 1)

    for f in demo_features:
        df[f] = df[f].replace({False: 0, True: 1})
    
    return df
```

```python
def residualize(input_df, brain_features):
    df = input_df.copy()

    mean_age = np.mean(df["interview_age"])

    confounders = ["interview_age"] + demo_features
    df[confounders] = df[confounders].fillna(0)
    
    for feature in brain_features:
        endog = df[feature]
        df["Intercept"] = 1
        
        exog = df[["Intercept"] + confounders]
        md = sm.GLM(endog, exog, family=sm.families.Gaussian())
    
        md = md.fit()
        #print(md.summary())
    
        feature_name = feature + '_prime'

        df[feature_name] = df[feature] - (md.params[1] * (df["interview_age"] - mean_age))
        for i in range(1, len(confounders)):
            df[feature_name] = df[feature_name] - (md.params[i + 1] * df[confounders[i]])

    return df
    
```

```python
baseline_c_stop_go_res = residualize(one_hot_encode(baseline_c_stop_go), c_stop_go_features)
followup_c_stop_go_res = residualize(one_hot_encode(followup_c_stop_go), c_stop_go_features)
```

```python
from sklearn.feature_selection import f_regression, SelectKBest

# Applying SelectKBest class to extract top 1 best features

NUM_FEATURES = 10

# set k to whatever number of features you want
fs = SelectKBest(score_func=f_regression,k=NUM_FEATURES)

# Applying feature selection
fit = fs.fit(X, y)

features_score = pd.DataFrame(fit.scores_)
features = pd.DataFrame(X.columns)
feature_score = pd.concat([features,features_score],axis=1)

# Assigning column names
feature_score.columns = ["Input_Features","F_Score"]
top_features = feature_score.nlargest(NUM_FEATURES,columns="F_Score")
top_features
```

```python
X_selected = baseline_c_stop_go_res[top_features["Input_Features"]]
y = baseline_c_stop_go_res["demo_sex_v2"]

scaler = StandardScaler().set_output(transform="pandas")
X_selected_scaled = scaler.fit_transform(X_selected)
```

## Standard Models

```python
names = [
     "SVC",
     "RandomForestClassifier",
     "KNeighborsClassifier",
     "MLPClassifier",
     "LogisticRegression"
]

standard_models = [
    SVC(),
    RandomForestClassifier(),
    KNeighborsClassifier(),
    MLPClassifier(hidden_layer_sizes=(16,32), max_iter=600, random_state=42),
    LogisticRegression(penalty="l2"),
]

def run_models(title, X_train, y_train):
    
    print(f"Running {title} experiment")
    print("=============================================")
    
    for name, model in zip(names, standard_models):
        print(f"Trying {name}...")
        model.fit(X_train, y_train)
        cv_scores = cross_val_score(model, X_train, y_train, cv=5)
        average_cv_score = cv_scores.mean()
        print(f"\tAverage score: {average_cv_score*100:.3f}%")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_selected_scaled, y, test_size=0.2, random_state=42)
```

## Connectome Analysis

```python
X_selected = baseline_c_stop_go_res[top_features["Input_Features"]]
y = baseline_c_stop_go_res["demo_sex_v2"]

scaler = StandardScaler().set_output(transform="pandas")
X_selected_scaled = scaler.fit_transform(X_selected)

print(X_selected_scaled.shape)
print(y.shape)
```

```python
X_selected_scaled.iloc[0].to_frame().transpose()
```

```python
males = X_selected_scaled[y == 0]
females = X_selected_scaled[y == 1]

males_corr = males.corr()
females_corr = females.corr()
diffmap = males_corr - females_corr

# visualize the mean control maps
fig = plt.figure(dpi=200)

gs = fig.add_gridspec(2,2)
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, :])

for ax in [ax1, ax2, ax3]:
    ax.grid(False)

ax1.set_title("Male Correlation Map")
plotting.plot_matrix(males_corr, vmax=0.8, vmin=-0.8, reorder=False, axes=ax1)
ax2.set_title("Female Correlation Map")
plotting.plot_matrix(females_corr, vmax=0.8, vmin=-0.8, reorder=False, axes=ax2)
ax3.set_title("Difference Correlation Map")
plotting.plot_matrix(diffmap, vmax=0.8, vmin=-0.8, reorder=False, axes=ax3)

plt.show()
```

```python
statistics, pvalues = ttest_ind(males_corr, females_corr, axis=0)

# Get adjusted p-values
p_adjusted = false_discovery_control(pvalues.flatten())

# Count significant p_values < 0.05
print("Correlations with p-values below a significance level of 0.05: ", sum(p_adjusted < 0.05))

# Find significant p_values < 0.05
np.argwhere(p_adjusted < 0.05)
```
